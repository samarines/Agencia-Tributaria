```python
import pandas as pd
df = pd.read_csv('../pandas/Agencia Tributaria/Descargas/modelo190_salarios(1).csv', sep=';')
df
```

```python
# Cambiar lo tipos
df.replace(',', '.', regex=True, inplace=True)
```

```python
# leyendas de cada una de las columnas
df_leyenda = pd.read_csv('/pandas/Agencia Tributaria/Modelo 190_salarios_leyenda_columnas.csv', sep=',')
df_leyenda
```

```python
# Extraigo los datos del año 2020, provincia de Santa Cruz de Tenerife, hombres

df_2020_hombres = df[(df['EJER'] == 2020) & (df['PROV'] == 38) & (df['SEXO'] == 1)]
df_2020_hombres
```

```python
# Extraigo los datos del año 2020, provincia de Santa Cruz de Tenerife, mujeres

df_2020_mujeres = df[(df['EJER'] == 2020) & (df['PROV'] == 38) & (df['SEXO'] == 2)]
df_2020_mujeres
```

# ¿Cuáles son los salarios por sectores?

```python
df_2020_hombres_sector = df_2020_hombres.iloc[:,[1,23]].astype(float)
df_2020_hombres_sector 
```

```python
df_simplificada_hombres = df_2020_hombres_sector.assign(Simplificada = lambda x: x.M190_2/1000000) 
df_simplificada_hombres
```

```python
df_2020_mujeres_sector = df_2020_mujeres.iloc[:,[1,23]].astype(float)
df_simplificada_mujeres = df_2020_mujeres_sector.assign(Simplificada = lambda x: x.M190_2/1000000)
df_simplificada_mujeres
```

```python
df_sectores = pd.read_csv('../pandas/Agencia Tributaria/Sector NACE.csv', sep=';')
df_sectores
```

```python
df_join_mujeres = df_simplificada_mujeres.set_index('SECTOR').join(df_sectores.set_index('SECTOR'))
df_join_mujeres
```

```python
import pandas
import matplotlib.pyplot as plt
import numpy as np

ind = np.arange(len(df_join_mujeres))


fig, ax = plt.subplots(figsize=(16,8))

width = 0.4
padding=0.2

ax.barh(ind, df_simplificada_mujeres.Simplificada, width, color='red', label='Mujeres')
ax.barh(ind+width, df_simplificada_hombres.Simplificada, width, color='green', label='Hombres')

ax.set(yticks=ind, yticklabels=df_join_mujeres[' Literal'])
ax.set_ylabel('Actividades')
ax.set_xlabel('Millones de euros')
ax.set_title('Salarios brutos por grupo de actividad. Provincia de Santa Cruz de Tenerife. Año 2020', fontsize=15)
ax.legend()


plt.show()
```
