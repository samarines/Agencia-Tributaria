Es difícil lidiar con los encabezados, sólo encuentro la solución de convertir a tuplas el marco de datos y reconstruirlo de nuevo.

### Gráfico de Pandas y Matplotlib

La Agencia publica uno doble, vamos a hacer uno de barras horizontales

```python
import numpy as np
import matplotlib.pyplot as plt

# El largo del marco es 12
# no quiero que tenga en cuenta la última línea(hay un Nan)
# tampoco la primera línea porque desfigura el gráfico por ser un total
N = 10
ind = np.arange(N)  
width = 0.35       

fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111)

ax.barh(ind, df_uno.iloc[1:11,4].astype(float).fillna(0), width, color='blue', label='Hombres')
ax.barh(ind+width, df_uno.iloc[1:11,5].astype(float).fillna(0), width, color='red', label='Mujeres')

ax.set_title('Asalariados por sectores económicos y género. Año 2020. Provincia de Santa Cruz de Tenerife.', fontsize=15)

ax.set_ylabel('Sectores')
ax.set(yticks=ind, yticklabels=df_uno.iloc[1:11,1])
ax.set_xticklabels(range(0,80,10), fontsize=13)
ax.set_xlabel('Asalariados (miles)')
ax.grid()
ax.legend()
```
